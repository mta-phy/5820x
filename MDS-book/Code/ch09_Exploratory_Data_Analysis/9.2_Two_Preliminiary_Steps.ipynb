Read the file `iris.csv` and prepare data. Results in the feature matrix X and target vector y:

```python
import numpy as np

data = np.loadtxt('iris.csv', dtype='str', delimiter=',')
X = np.array(data[:, :4], dtype=float)
y = np.array(data[:, 4], dtype=str)

print(f"X has {X.shape[0]} rows and {X.shape[1]} columns.")
```
