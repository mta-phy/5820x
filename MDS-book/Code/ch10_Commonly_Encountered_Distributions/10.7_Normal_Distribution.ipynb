```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats

from utilities import pmfplot
```

## 10.7 Normal Distribution

### Plot the Continuous Distribution

```python
xlim = (-8, 8)

x = np.linspace(*xlim, 200)
mu = 1.5
sigma = 2

rv = stats.norm(loc=mu, scale=sigma)

fig, ax = plt.subplots(nrows=2, sharex=True, figsize=(4, 4.5))
ax[0].plot(x, rv.pdf(x))
ax[1].plot(x, rv.cdf(x))
ax[0].set(ylabel=r"PDF $f(x)$", xlim=xlim, ylim=(-0.01, 0.25))
ax[1].set(xlabel="random variable $x$", ylabel=r"CDF $F(x)$", 
          xlim=xlim, ylim=(-0.02, 1.1),
          xticks=range(-8, 9, 2))    

print(f"variance ={rv.var():5.3},  std. dev ={rv.std():5.3},  mean ={rv.mean():5.3},  median ={rv.median():5.3}")
```

### Draw Samples from the Distribution
The randomly drawn sample is shown as the black marker below the line plot. The marker are partially opaque which results in a darker color in regions where they overlap (and thus have a higher density).

```python
x = np.linspace(-4, 0., 200)
mu, sigma = -2, 0.5  # the two parameter of the normal distribution

# create a "random variable" object for the normal distribution
rv = stats.norm(loc=mu, scale=sigma)

fig, ax = plt.subplots(nrows=2, sharex=True, gridspec_kw={'height_ratios': [1, 0.2]})
ax[0].plot(x, rv.pdf(x))
ax[0].set(xlabel="random variable $x$", ylabel='PDF')

# draw 100 random numbers based on a normal distributed probability
r = rv.rvs(100)
ax[1].plot(r, np.zeros_like(r), 'o', alpha=0.4, markersize=8, mfc='k', mec='none') 
ax[1].set_axis_off()
```
