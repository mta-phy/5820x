```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
```

## 10.13 Logistic Distribution

### Plot the Continuous Distribution

Additionally, we superimposed as dashed line a normal distribution with the best-fitting parameters. This helps to see the difference between these two functions:

```python
xlim = (-4, 4)

x = np.linspace(*xlim, 200)
mu = 0
sigma = 0.5

rv = stats.logistic(loc=mu, scale=sigma)

fig, ax = plt.subplots(nrows=2, sharex=True, figsize=(4, 4.5))
ax[0].plot(x, rv.pdf(x))
ax[1].plot(x, rv.cdf(x), label='logistic')
ax[0].set(ylabel=r"PDF $f(x)$", xlim=xlim, ylim=(-0.01, 0.6))
ax[1].set(xlabel="random variable $x$", ylabel=r"CDF $F(x)$", 
          xlim=xlim, ylim=(-0.02, 1.1))    

print(f"variance ={rv.var():5.3},  std. dev ={rv.std():5.3},  mean ={rv.mean():5.3},  median ={rv.median():5.3}")


rv_normal = stats.norm(loc=rv.mean(), scale=rv.std())
ax[0].plot(x, rv_normal.pdf(x), ls='--', c='C3')
ax[1].plot(x, rv_normal.cdf(x), ls='--', c='C3', label='best-fitting normal')

ax[1].legend();
```
