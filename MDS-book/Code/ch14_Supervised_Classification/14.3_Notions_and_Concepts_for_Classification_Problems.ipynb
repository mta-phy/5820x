## 14.3 Notions and Concepts for Classification Problems
### 14.3.1 Datasets, Models, and the Hypothesis Class
We use the iris dataset from the package `mdsdata`(see https://mds-book.org/Content/datasets)

```python
from mdsdata import DS1
```

```python
DS = DS1.load_data()
X1 = DS.data[:,3]
X2 = DS.data[:,2]
y = DS.target
```

```python
mask = (y == 0)
x0 = np.min(X1[mask])
x1 = np.max(X1[mask])
y0 = np.min(X2[mask])
y1 = np.max(X2[mask])
```

Here is an alternative usage of the datasets (the simplified interface) which might be somewhat more intuitive:

```python
from mdsdata import load_iris
_, _, X2, X1, Y = load_iris()  # the first 4 returned values are the four features

mask = (Y == 0)
x0 = np.min(X1[mask])
x1 = np.max(X1[mask])
y0 = np.min(X2[mask])
y1 = np.max(X2[mask])
```

```python
fig, ax = plt.subplots()
ax.plot([x0, x1, x1, x0, x0], [y0, y0, y1, y1, y0], c='C1')
ax.scatter(X1, X2, c=Y);
```
