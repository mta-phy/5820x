## 8.6 Central Limit Theorem

Assume that a number of $n$ random experiments are performed, for each of which
the mean value $\bar{ùë•}_i$ of the random variable $X$ is recorded. Then, the central limit
theorem (CLT) states that the shape of the distribution of the $n$ values of the sample
means, $\bar{ùë•}_i$, always approaches the shape of a Gaussian distribution.

Example: An experiment of sampling integers from a uniform
random distribution with values in between -10 and 10, all of which have the same likelihood to appear.

```python
import numpy as np

def gaussian(x, loc, s):
    return (1 / np.sqrt(2 * np.pi * s ** 2) *
            np.exp(-(x - loc) ** 2 / (2 * s ** 2)))
```

```python
# larger values = narrower distributions
# If this is changed, then also the parameters for the Gaussian, plotted as line below
# needs to be changed!
sample_size = 10000


# larger values = more "Gaussian-like"
# Try to change this value and observe that for larger values the distribution of the means
# becomes more and more Gaussian-like
n_means = 800
```

```python
means = []
for i in range(n_means):
    sample = np.random.randint(-10, 11, size=sample_size)
    sample_mean = np.mean(sample)
    means.append(sample_mean)
```

```python
import matplotlib.pyplot as plt
plt.hist(means, bins=30, density=True)
x = np.linspace(-0.4, 0.4, 200)
plt.plot(x, gaussian(x, 0, 0.06))  # manually "fitting" a Gaussian to the data
```
