```python
import numpy as np
X=[-2,-1,0,1,2]
Y=[5,2,4, 1,3]
covariance=np.cov(X,Y)
print("covariance matrix\n", covariance)
pearson=np.corrcoef(X, Y)
print("pearson correlation coefficient=", np.round(pearson[0,1], decimals=1))
```
