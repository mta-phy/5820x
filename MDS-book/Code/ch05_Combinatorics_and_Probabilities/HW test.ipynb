# Excerise 5.11

```python
np.random.seed(40)
n=20; N=100
samples = np.random.randint(1, N+1, n)

event_A = 0 
event_B = 0 
event_A_and_B = 0 

for i in samples:
    if i < 50:
        event_A+=1
    if i%2==1:
        event_B+=1
    if (i<50) & (i%2 == 1):
        event_A_and_B+=1

print("Samples:", samples)
print("P(A):", event_A/n)
print("P(B):", event_B/n)
print("P(A ∩ B):", event_A_and_B/n)

print('\nTheoretical Probabilites:\nLikelihood of event A = ' + str(pA))
print('Likelihood of event B = ' + str(pB))
print('Likelihood of both event A & B = ' + str(pA_B))
```

# Excerise 6.5

```python
# (a)
import numpy as np
import matplotlib.pyplot as plt

k_B = 1.3806e-23  # in J/K
T = 298           # in Kelvin
amu_to_kg = 1.6605e-27
m = 4 * amu_to_kg  # Mass of helium atom in kg

# Maxwell-Boltzmann speed distribution function
def f_MB(c, m, T):
    factor = 4 * np.pi * c**2 * (m / (2 * np.pi * k_B * T))**(3/2)
    exponent = np.exp(-m * c**2 / (2 * k_B * T))
    return factor * exponent


c = np.linspace(0, 3000, 500)
f_c = f_MB(c, m, T)

# Plotting
plt.figure(figsize=(8, 5))
plt.plot(c, f_c, label='Helium at 298 K')
plt.xlabel('Speed (m/s)')
plt.ylabel('Probability Density Function for He')
plt.title('Maxwell-Boltzmann Distribution of Speeds')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
