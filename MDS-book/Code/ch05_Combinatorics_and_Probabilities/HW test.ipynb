```python
def cumulative_sum(List):
    com_sum=[]
    sum=0
    for i in List:
        com_sum.append(sum+i)
        sum=sum+i
    return com_sum
mylist=[1,2,3,4,5,6]
result_1=cumulative_sum(mylist)

import numpy as np
result_2=np.cumsum(mylist)
print('cumulative sum using defined function = ' + str(result_1) + '\ncumulative sum using numpy.cumsum() = ' + str(result_1))
```

# Excerise 5.11

```python
np.random.seed(40)
n=20; N=100
samples = np.random.randint(1, N+1, n)

event_A = 0 
event_B = 0 
event_A_and_B = 0 

for i in samples:
    if i < 50:
        event_A+=1
    if i%2==1:
        event_B+=1
    if (i<50) & (i%2 == 1):
        event_A_and_B+=1

print("Samples:", samples)
print("P(A):", event_A/n)
print("P(B):", event_B/n)
print("P(A ∩ B):", event_A_and_B/n)

print('\nTheoretical Probabilites:\nLikelihood of event A = ' + str(pA))
print('Likelihood of event B = ' + str(pB))
print('Likelihood of both event A & B = ' + str(pA_B))
```

```python
import numpy as np

def simulate_events(n, N, seed=None):
    if seed is not None:
        np.random.seed(seed)

    samples = np.random.randint(1, N+1, size=n)

    event_A = samples < 50
    event_B = samples%2 == 1
    event_A_and_B = event_A & event_B

    pA = event_A.mean()
    pB = event_B.mean()
    pAB = event_A_and_B.mean()

    return samples, pA, pB, pAB


samples, pA, pB, pAB = simulate_events(n=20, N=100, seed=40)

print("Samples:", samples)
print("P(A):", pA)
print("P(B):", pB)
print("P(A ∩ B):", pAB)
```
